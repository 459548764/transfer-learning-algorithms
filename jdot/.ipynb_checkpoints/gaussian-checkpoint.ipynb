# JDOT Domain Adaptation with Gaussian Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import keras
```

```python
colormap = np.array(plt.rcParams['axes.prop_cycle'].by_key()['color'])
```

## Create the dataset

### Source domain

```python
N_source = [50, 50] # number of points for the 2 classes
```

```python
mean = np.array([[-1,0], [1,0]])
std = 0.1
cov = np.eye(2) * std
```

```python
X = []
for i_class in range(len(N_source)):
    x, y = np.random.multivariate_normal(mean[i_class], cov, N_source[i_class]).T
    X.append(np.concatenate([x.reshape(len(x), 1), y.reshape(len(y), 1)], axis=1))
X_source = np.concatenate([X[i_class] for i_class in range(len(N_source))], axis=0)
Y_source = np.concatenate([[i_class]*N_source[i_class] for i_class in range(len(N_source))])
```

### Target domain

```python
N_target = [50, 50] # number of points for the 2 classes
```

```python
mean = np.array([[-0.7,-0.7], [0.7,0.7]])
std = 0.1
cov = np.eye(2) * std
```

```python
X = []
for i_class in range(len(N_target)):
    x, y = np.random.multivariate_normal(mean[i_class], cov, N_target[i_class]).T
    X.append(np.concatenate([x.reshape(len(x), 1), y.reshape(len(y), 1)], axis=1))
X_target = np.concatenate([X[i_class] for i_class in range(len(N_target))], axis=0)
Y_target = np.concatenate([[i_class]*N_target[i_class] for i_class in range(len(N_target))])
```

### Display the dataset

```python
plt.rcParams['figure.figsize'] = (8, 6)
plt.axis("equal")
plt.axis([-2,2,-2,2])

plt.scatter(X_source[:,0], X_source[:,1], label="source", s=40, facecolors='none', edgecolor=colormap[Y_source])
plt.scatter(X_target[:,0], X_target[:,1], label="target", color=colormap[Y_target])
plt.legend()
```

## Initialization

### Model for the classifier

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(1,activation='sigmoid', input_dim=2))
```

```python
model.compile('sgd','mse')
```

```python
model.fit(X_source, Y_source, epochs=5, batch_size=20)
```

```python
plt.rcParams['figure.figsize'] = (8, 6)
plt.axis("equal")
plt.axis([-2,2,-2,2])

x_min, x_max = -4, 4
y_min, y_max = -4, 4
h = 0.1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
levels = np.linspace(0,1,20)
cm = plt.cm.coolwarm
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
cont = plt.contourf(xx, yy, Z, levels, cmap=cm, alpha=.8)
plt.colorbar(cont, orientation="vertical")

plt.title("Classifier of the source domain (hollow points)")
plt.scatter(X_source[:,0], X_source[:,1], label="source", s=40, facecolors='none', edgecolor=colormap[Y_source])
plt.scatter(X_target[:,0], X_target[:,1], label="target", color=colormap[Y_target])
plt.legend()
```
